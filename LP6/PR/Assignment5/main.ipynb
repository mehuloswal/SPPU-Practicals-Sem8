```python
# import numpy as np
# import matplotlib.pyplot as plt
# from skimage.io import imread, imsave
# from skimage.color import rgb2gray
# from skimage.filters import threshold_otsu
# from skimage.segmentation import watershed
# from skimage.measure import label, regionprops
# from skimage.feature import peak_local_max
# from scipy import ndimage as ndi

# # Load an image from the Berkeley Segmentation Dataset
# img = imread('/home/burhanuddin/College/BE/sem_8/labs/LP6/assignment_5/archive/images/test/2018.jpg')

# # Convert the image to grayscale
# img_gray = rgb2gray(img)

# # Threshold the image using Otsu's method
# thresh = threshold_otsu(img_gray)
# img_binary = img_gray > thresh

# # Apply watershed segmentation
# distance = ndi.distance_transform_edt(img_binary)
# local_maxi = peak_local_max(distance, footprint=np.ones((3, 3)), labels=img_binary)
# markers = label(local_maxi)[0]
# labels = watershed(-distance, markers, mask=img_binary)

# # Compute region properties for each segmented region
# props = regionprops(labels)

# # Visualize the segmentation results
# fig, ax = plt.subplots(figsize=(10, 10))
# ax.imshow(img)
# for prop in props:
#     y0, x0 = prop.centroid
#     orientation = prop.orientation
#     x1 = x0 + np.cos(orientation) * 0.5 * prop.major_axis_length
#     y1 = y0 - np.sin(orientation) * 0.5 * prop.major_axis_length
#     x2 = x0 - np.sin(orientation) * 0.5 * prop.minor_axis_length
#     y2 = y0 - np.cos(orientation) * 0.5 * prop.minor_axis_length
#     ax.plot((x0, x1), (y0, y1), '-r', linewidth=2.5)
#     ax.plot((x0, x2), (y0, y2), '-r', linewidth=2.5)
#     ax.plot(x0, y0, '.g', markersize=15)

# # Save the segmentation result
# imsave('/home/burhanuddin/College/BE/sem_8/labs/LP6/assignment_5/archive/2018.jpg', labels)

import numpy as np
import matplotlib.pyplot as plt
from skimage import io, filters, morphology, segmentation
from skimage.measure import label, regionprops
from skimage.feature import peak_local_max
from scipy import ndimage

# Load image
image = io.imread('https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg')

# Convert to grayscale
image_gray = np.mean(image, axis=2)

# Apply Gaussian filter to remove noise
image_smooth = filters.gaussian(image_gray, sigma=1)

# Find edges using the Sobel operator
edges = filters.sobel(image_smooth)

# Threshold edges to obtain binary image
edges_binary = edges > filters.threshold_otsu(edges)

# Morphological opening to remove small objects
edges_binary = morphology.binary_opening(edges_binary, morphology.disk(3))

# Compute Euclidean distance transform
distance = ndimage.distance_transform_edt(edges_binary)

# Find local maxima in distance transform
local_maxi = peak_local_max(distance, footprint=np.ones((3, 3)), labels=edges_binary)
markers = label(local_maxi)[0]

# Apply watershed algorithm to segment regions
labels = segmentation.watershed(-distance, markers, mask=edges_binary)

# Compute region properties for each segmented region
props = regionprops(labels)

# Plot original image and segmented regions
fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].imshow(image)
ax[1].imshow(segmentation.mark_boundaries(image, labels))
plt.show()
```
