*bootstrap *- Allows for resampling of the training dataset without replacement.

oob_score **- Used to compute the model’s accuracy score after training.bold text**

```python
from sklearn.preprocessing import StandardScaler
x_std=StandardScaler().fit_transform(x)
```

```python
x_std
```

```python
X1=x_std.T
X1
```

```python
#covariance matrix
cov_mat=np.cov(x_std.T)
cov_mat
```

```python
#eigen value and eigen vector
eig_vals,eig_vecs=np.linalg.eig(cov_mat)
eig_vals
```

```python
eig_vecs
```

```python
#feature vector
#eplained variance
tot=sum(eig_vals)
var_exp=[(i/tot)*100 for i in sorted(eig_vals,reverse=True)]
var_exp
```

```python
cum_var_exp=np.cumsum(var_exp)
cum_var_exp
```

```python
plt.figure(figsize=(10,5))
plt.bar(range(len(var_exp)),var_exp, label='indivisual explained variance',color='g')
plt.step(range(len(cum_var_exp)),cum_var_exp,label='cummulative explained variance')
plt.ylabel('explained variance ratio')
plt.xlabel('principal compoents')
plt.legend()
plt.show()
```
